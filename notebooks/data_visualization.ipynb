# 📊 Luxury Market Trends Dashboard
Exploratory Data Analysis for luxury real estate listings in SW Florida

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('../data/luxury_real_estate_data.csv')
df['listing_date'] = pd.to_datetime(df['listing_date'])
df.head()
```

```python
# Example: Avg price per sqft over time
df['month'] = df['listing_date'].dt.to_period('M').astype(str)
avg_price_trend = df.groupby(['month', 'zip_code'])['price_per_sqft'].mean().reset_index()
fig = px.line(avg_price_trend, x='month', y='price_per_sqft', color='zip_code', title='Price per Sqft Over Time')
fig.show()
```
